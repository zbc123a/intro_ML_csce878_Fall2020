```python
def polynomialFeatures(X, degree=1):
    def poly_array(x):
        poly_combination = []
        for i in range(degree):
            poly_combination.append(combinations_with_replacement(x, i+1))

        poly = []
        for i in poly_combination:
            for j in i:
                poly.append(j)

        arr = []
        for a in poly:
            result = 1
            for b in a:
                index = np.where(x==b)[0][0]
                result = result * x[index]
            arr.append(result)
        return arr
    count = 0
    if X.ndim >1:
        for row in X:
            a = poly_array(row)
            if(count==0):
                feature_array = a
            else:
                feature_array = np.vstack((feature_array, a))
            count += 1
        return np.array(np.insert(feature_array, 0, 1, axis=1))
    else:
        return np.array(np.insert(poly_array(X), 0, 1, axis=1))
```
